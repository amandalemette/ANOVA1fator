####  Somas quadráticas --> SQ

- SQ_tratamentos:

```python
ytt = dados['médias'][5]
ytt
```

```python
dados['médias'][0:5]
```

```python
n = 5
a = 5
```

```python
SQ_trat = np.sum((dados['médias'][0:5]-ytt)**2)*n
SQ_trat
```

```python
GL_trat = a-1
GL_trat
```

- SQ_E:

```python
dados.loc[0][1:6]
```

```python
dados['médias'][0]
```

```python
(dados.loc[0][1:6]-dados['médias'][0])**2
```

```python
dados
```

```python
SQ_E = np.sum([(dados.loc[j][1:6]-dados['médias'][j])**2 for j in range (0,5)])
SQ_E
```

####  Graus de liberdade --> GL

```python
GL_E = a*(n-1)
GL_E
```

```python
SQ_T = SQ_trat + SQ_E
SQ_T
```

```python
GL_T = GL_trat + GL_E
GL_T
```

####  Médias Quadráticas --> MQ

```python
MQ_trat = SQ_trat/GL_trat
MQ_trat
```

```python
MQ_E = SQ_E/GL_E
MQ_E
```

#### Cálculo de F:

```python
F = MQ_trat/MQ_E
F
```

```python
import scipy.stats
```

```python
alpha = 0.01
```

- Determinação do F tabelado:

```python
F_tab = scipy.stats.f.ppf(q = 1-alpha, dfn = GL_trat,dfd = GL_E)
F_tab
```

- Determinação do p-valor:

```python
p_valor = scipy.stats.f.sf(F,GL_trat,GL_E)
p_valor
```

- Tabela da ANOVA

```python
anova = pd.DataFrame({'Fonte de Variação':['Conc. algodão','Erro', 'Total'],
                     'SQ':[SQ_trat,SQ_E,SQ_T],'GL':[GL_trat,GL_E,GL_T],'MQ':[MQ_trat,MQ_E,''],
                     'F':[F,'',''],'p-valor':[p_valor,'','']})
```

```python
anova
```

- Como $F_{calculado} > F_{tabelado}$ e $p_{valor} < \alpha$, a hipótese nula é rejeitada e a concentração de algodão exerce sim efeito estatisticamente significativo na resistência do tecido sintético
